Equations System.

```python
eq1 = x + 2*y - 4
eq2 = x - y + 2
```

```python
sympy.solve([eq1, eq2], [x, y], dict=True)
```

```python
eq1 = x**2 - y
eq2 = y**2 - x
```

```python
sols = sympy.solve([eq1, eq2], [x, y], dict=True)
sols
```

checking the roots

```python
[eq1.subs(sol).simplify() == 0 and eq2.subs(sol).simplify() == 0 for sol in sols]
```
